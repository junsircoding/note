```python
from tensorflow import keras as kr
```

```python
# 创建神经网络
model = kr.Sequential([
    kr.layers.Dense(units=1, input_shape=[1])
])
```

```python
# 编译神经网络
model.compile(optimizer="sgd", loss="mean_squared_error")
```

```python
import numpy as np
# 构造数据和答案
data_x = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
answer_y = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

```python
# 训练神经网络
model.fit(data_x, answer_y, epochs=500)
```

```python
# 进行预测
new_data = [10.0]
result = model.predict(new_data)
result
```

```python
# 保存模型
model.save("/tmp/test/")
```
