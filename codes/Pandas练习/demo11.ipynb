```python
import pandas as pd


class Solution:
    def getIndex(self, k, m, deliciousness):
        result = -1
        len_d = len(deliciousness)
        de_rev = deliciousness[-1::-1]
        ree = []
        for i in range(len_d):
            if de_rev[i] > m:
                ree.append(len_d-i-1)
        if len(ree) >= k:
            result = ree[k-1]
        return result

    def getPosition(self, m, k, number):
        p = -1
        pl = []
        for i in range(0, len(number)):
            print(number[0:i+1])
            he = sum(number[0:i+1])
            if he % m == k:
                pl.append(i)
        print(pl)
        if pl:
            p = max(pl)
        return p


so = Solution()
print(so.getPosition(
    m=9,
    k=6,
    number=[6, 3, 4, 9, 10, 7, 10, 2, 6]
    # m = 6,
    # k = 0,
    # number=[3,1,3,2,2,2]
    # -1
    # k=1,
    # m=3,
    # deliciousness=[6, 9, 3, 6]
))
# 3


df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth"
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"]
    }
)
```
