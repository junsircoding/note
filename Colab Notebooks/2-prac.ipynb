```python
import tensorflow as tf
```

```python
# TODO: 实现一个加法运算
# 创建两个常量
con_a = tf.constant(3.0)
con_b = tf.constant(4.0)
# 调用加法方法
sum_c = tf.add(con_a, con_b)
print(f"con_a: {con_a}")
print(f"con_b: {con_b}")
print(f"sum_c: {sum_c}")
```

```python
# 获取默认图
print(f"默认图: {tf.get_default_graph()}")
print(f"con_a 的图属性为: {con_a.graph}")
print(f"con_b 的图属性为: {con_b.graph}")
print(f"sum_c 的图属性为: {sum_c.graph}")
```

```python
# tf 的变量和操作只能通过会话执行, 若不指定图, 则运行默认的图
with tf.Session() as sess:
  print(f"con_a: {sess.run(con_a)}")
  print(f"con_b: {sess.run(con_b)}")
  print(f"sum_c: {sess.run(sum_c)}")
  print(f"该会话的图属性为: {sess.graph}")
```

```python
# 创建图
new_graph = tf.Graph()
with new_graph.as_default():
  # 在图中创建了一个 operation
  con_d = tf.constant(30.0)
  print(f"con_d 的属性为: {con_d.graph}")

with tf.Session(graph=new_graph) as new_sess:
  print(f"con_d 的值是: {new_sess.run(con_d)}")
  print(f"新会话的图属性是: {new_sess.graph}")
```
